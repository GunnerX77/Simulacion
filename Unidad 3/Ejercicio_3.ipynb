```python
import numpy as np
import matplotlib.pyplot as plt

def simular_tiempos_espera(clientes=50, media=10, desviacion=2):
    # Convertimos media y desviación estándar a los parámetros que tendrá la distribución lognormal:
    mu = np.log(media**2 / np.sqrt(media**2 + desviacion**2))
    sigma = np.sqrt(np.log(1 + (desviacion**2 / media**2)))

    # Generamos los tiempos de espera usando la distribución lognormal:
    tiempos_espera = np.random.lognormal(mean=mu, sigma=sigma, size=clientes)

    # Mostramos los resultados
    print("Tiempos de espera para 50 clientes:", tiempos_espera)
    print("Tiempo máximo de espera registrado:", np.max(tiempos_espera))

    # Mi gráfica:
    plt.hist(tiempos_espera, bins=15, color='orange', edgecolor='black', density=True)
    plt.xlabel('Tiempo de espera (minutos)')
    plt.ylabel('Frecuencia relativa')
    plt.title('Distribución LogNormal de Tiempos de Espera')
    plt.show()

# Ejecutamos la función:
simular_tiempos_espera()
```
