```python
import numpy as np
import matplotlib.pyplot as plt

def simular_pedidos_satisfechos(dias=7, pedidos_por_dia=20, probabilidad_exito=0.7):
    # Generamos datos usando la distribuci√≥n binomial:
    pedidos_satisfechos = np.random.binomial(n=pedidos_por_dia, p=probabilidad_exito, size=dias)

    # Mostramos resultados:
    print("Pedidos satisfechos por d√≠a:", pedidos_satisfechos)

    # Grafica de los resultados:
    plt.bar(range(1, dias + 1), pedidos_satisfechos, color='skyblue')
    plt.xlabel('Dias')
    plt.ylabel('Pedidos Cumplidos')
    plt.title('Simulaci√≥n de Pedidos Cumplidos por D√≠a')
    plt.xticks(range(1, dias + 1))
    plt.show()

# Ejecutar simulaci√≥n
simular_pedidos_satisfechos()
```
